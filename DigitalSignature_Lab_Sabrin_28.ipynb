```python
from ecdsa import SigningKey, NIST256p
import hashlib

# Transaction message
message = "Ahmed pays 100 coins to Ali"
print("Transaction Message:", message)

# Hashing the message using SHA-256
message_hash = hashlib.sha256(message.encode()).digest()
print("SHA-256 Hash:", message_hash.hex())

# Generating private and public keys using NIST256p curve
private_key = SigningKey.generate(curve=NIST256p)
public_key = private_key.get_verifying_key()

# Signing the hash with the private key
signature = private_key.sign(message_hash)
print("Signature:", signature.hex())

# Verifying the signature using the public key
verified = public_key.verify(signature, message_hash)
print("Signature Verified:", verified)
```
