```python
p = 5
q = 11
n = p * q
phi = (p - 1) * (q - 1)
e = 3  # must be coprime with phi
d = 27  # private key, since (e * d) % phi = 1

# Encryption and Decryption functions
def encrypt(m):
    return pow(m, e, n)

def decrypt(c):
    return pow(c, d, n)

#  Two plaintext messages to encrypt
m1 = 7
m2 = 9

c1 = encrypt(m1)
c2 = encrypt(m2)

print("Encrypted m1:", c1)
print("Encrypted m2:", c2)

# خاصية التلازم الضربي
homomorphic_product = (c1 * c2) % n
combined_message = (m1 * m2) % n
encrypted_combined = encrypt(combined_message)

print("Product of ciphertexts:", homomorphic_product)
print("Encryption of product:", encrypted_combined)
print("Homomorphism holds:", homomorphic_product == encrypted_combined)

# Decrypt the product to verify the result
print("Decrypt product:", decrypt(homomorphic_product))
```
